# Cleaning of the data

```python
# Convert 'MP' (minutes played) to float (minutes)
def convert_mp(mp):
    try:
        h, m, s = map(int, mp.split(':'))
        return h * 60 + m + s / 60
    except:
        return None

df['MP'] = df['MP'].apply(convert_mp)

# Convert numeric columns
for col in df.columns:
    if df[col].dtype == 'object':
        try:
            df[col] = pd.to_numeric(df[col])
        except:
            pass # Ignore conversion errors
```

```python
# Cleaning the 'Result' column
df['WinFlag'] = df['Result'].str.extract(r'^(W|L)') # Extract W or L
df['IsWin'] = df['WinFlag'] == 'W' # win boolean
df['IsWin'] = df['IsWin'].astype(int) # Convert boolean to int

score_pattern = r'(\d+)-(\d+)'  # Captures '123-116' as ('123', '116')
df[['TeamScore', 'OppScore']] = df['Result'].str.extract(score_pattern).astype(float) # Extract the scores
df['PointDiff'] = df['TeamScore'] - df['OppScore'] # Point differential

# Preview
print(df[['Result', 'TeamScore', 'OppScore', 'IsWin', 'PointDiff']].head())
```

```python
# Remove unnecessary columns
df = df.drop(columns=['Unnamed: 5', 'Rk', 'Gcar', 'Gtm', 'Result'], errors='ignore')

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

df.dropna(subset=['MP'], inplace=True)  # Drop rows where 'MP' is NaN
df = df.fillna(df.median(numeric_only=True)) # Fill remaining NaNs with median
```

# Team Data

```python
# Directory and filename pattern
base_dir = 'Data/SecondaryDataset-BasketballReference/Team Standings'
seasons = [f'{y}-{y+1}' for y in range(2019, 2025)]

# Columns to rename (if they exist in the file)
rename_map = {
    'Unnamed: 0': 'Rank',
    'Unnamed: 1': 'Team',
    'Unnamed: 2': 'Overall',
    'Place': 'Home',
    'Unnamed: 4': 'Road',
    'Conference': 'Conf_Record',
    'Unnamed: 6': 'Div_Record',
    'Margin': 'Point_Margin'
}

# Store individual DataFrames by season
cleaned_data = []

for season in seasons:
    file_path = os.path.join(base_dir, f'Team_standings_{season}.csv')
    df2 = pd.read_csv(file_path, header=1)
    df2.rename(columns={k: v for k, v in rename_map.items() if k in df2.columns}, inplace=True)
    df2['Season'] = season  # Add season column
    cleaned_data.append(df2)
    print(f"Cleaned '{season}' with columns: {df2.columns.tolist()}")

# Combine all into a single DataFrame
teams = pd.concat(cleaned_data, ignore_index=True)
```

```python
# Month abbreviation to number string
month_map = {'Jan': '01', 'Feb': '02', 'Mar': '03','Apr': '04', 'May': '05', 'Jun': '06'
             ,'Jul': '07', 'Aug': '08', 'Sep': '09','Oct': '10', 'Nov': '11', 'Dec': '12'}

# Columns to clean
target_columns = ['Home', 'Road', 'E', 'W', 'A', 'C', 'SE', 'NW', 'P', 'SW','Pre', 'Post', '?3', '?10','Oct', 'Nov', 'Dec', 'Jan', 'Feb', 'Mar','Apr','May','Jul','Aug']

for col in set(teams.columns) & set(target_columns):
    teams[col] = teams[col].astype(str).replace(month_map, regex=True) # This fixes the formatting that excel changes
```

```python
# Check for missing values
print("\nMissing Values in Team Standings Data:")
print(teams.isnull().sum())
teams.dropna(subset=['Rk', 'Team'], inplace=True)  # Drop rows where 'Rk' or 'Team' is NaN
```

```python
teams['Wins'] = teams['Overall'].str.extract(r'(\d+)-')[0].astype(int) # Extract wins from Overall column
teams['Losses'] = teams['Overall'].str.extract(r'-(\d+)')[0].astype(int) # Extract losses from Overall column
teams['WinPercentage'] = teams['Wins'] / (teams['Wins'] + teams['Losses']) # Calculate win percentage
teams['WinPercentage'] = teams['WinPercentage'].round(3) # Round to 3 decimal places
teams['WinPercentage'] = teams['WinPercentage'].astype(float) # Convert to float
```

```python
filtered_df = teams[teams['Team'].isin(['Los Angeles Lakers', 'Dallas Mavericks'])]
filtered_df = filtered_df.sort_values(by='Season')
pivot_df = filtered_df.pivot(index='Season', columns='Team', values='Wins') # Pivot the table so Seasons are index and teams are columns

# Plot
plt.figure(figsize=(10, 6))
plt.plot(pivot_df.index, pivot_df['Los Angeles Lakers'], marker='o', label='Lakers', color='gold')
plt.plot(pivot_df.index, pivot_df['Dallas Mavericks'], marker='o', label='Mavericks', color='blue')

plt.title('Wins by Season: Lakers vs Mavericks (Last 5 Seasons)')
plt.xlabel('Season')
plt.ylabel('Wins')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Feature Selection

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import VarianceThreshold, SequentialFeatureSelector, RFE
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Drop non-numeric and identifier columns
df_model = df.copy()
drop_cols = ["Date", "Team", "Opp", "GS", "Season", "WinFlag", "TeamScore", "OppScore", "IsWin"]
df_model = df_model.drop(columns=drop_cols)

# For comparing easier we'll be using the entire dataset here (as opposed of one for each player)
X = df_model.drop(columns=["PointDiff", "Player"])
y = df_model["PointDiff"]
```

```python
# Low variance filter
vt = VarianceThreshold(threshold=0.01)  # Default is 0.0
X_vt = vt.fit_transform(X)
vt_features = X.columns[vt.get_support()].tolist()[:3]  # Top 3
```

```python
# Recursive Feature Elimination (RFE)
rfe = RFE(estimator=LinearRegression(), n_features_to_select=3)
rfe.fit(X, y)
rfe_features = X.columns[rfe.support_].tolist()
```

```python
# Sequential Feature Selector (SFS)
sfs = SequentialFeatureSelector(LinearRegression()
                                , n_features_to_select=3
                                , direction='forward')
sfs.fit(X, y)
sfs_features = X.columns[sfs.get_support()].tolist()
```

```python
# Combine and compare
comparison_df = pd.DataFrame({
    'VarianceThreshold': pd.Series(vt_features),
    'RFE': pd.Series(rfe_features),
    'SequentialFeatureSelector': pd.Series(sfs_features)
})

print("Top 3 features selected by each method:")
print(comparison_df)
```

# Models

```python
# Filter for each player
df_player_77 = df_model[df_model["Player"] == "Luka Doncic"].copy()
df_player_23 = df_model[df_model["Player"] == "LeBron James"].copy()
```

```python
# Define features and target variables
X_77 = df_player_77.drop(columns=["PointDiff", "Player"])
X_23 = df_player_23.drop(columns=["PointDiff", "Player"])
y_77 = df_player_77["PointDiff"]
y_23 = df_player_23["PointDiff"]

# Apply Variance Threshold filter to each df
selector = VarianceThreshold(threshold=0.2) # Arbitrary 20%
X_reduced_77 = selector.fit_transform(X_77)
X_reduced_23 = selector.fit_transform(X_23)
feature_mask = selector.get_support(indices=True)
feature_names_77 = X_77.columns[feature_mask]
feature_names_23 = X_23.columns[feature_mask]
```

```python
# Train-test split
X_train_77, X_test_77, y_train_77, y_test_77 = train_test_split(X_reduced_77, y_77, test_size=0.2, random_state=42)
X_train_23, X_test_23, y_train_23, y_test_23 = train_test_split(X_reduced_23, y_23, test_size=0.2, random_state=42)

# Fit Linear Regression model
model_77 = LinearRegression()
model_23 = LinearRegression()
model_77.fit(X_train_77, y_train_77)
model_23.fit(X_train_23, y_train_23)
y_pred_77 = model_77.predict(X_test_77)
y_pred_23 = model_23.predict(X_test_23)

# Evaluate
mse_77 = mean_squared_error(y_test_77, y_pred_77)
print(f"Mean Squared Error for Luka's model: {mse_77:.2f}")
mse_23 = mean_squared_error(y_test_23, y_pred_23)
print(f"Mean Squared Error for Lebron's model: {mse_23:.2f}")
mse_77 = r2_score(y_test_77, y_pred_77)
print(f"R2 for Lukas's model: {mse_77:.2f}")
mse_23 = r2_score(y_test_23, y_pred_23)
print(f"R2 for Lebron's model: {mse_23:.2f}")
```

```python
# Store models and feature names in a dictionary
models = {
    "Luka Doncic": (model_77, feature_names_77),
    "LeBron James": (model_23, feature_names_23)
}

# Extract coefficients and variables for both players
coefficients_data = []

for player, (model, feature_names) in models.items():
    coefficients_data.extend(
        zip([player] * len(model.coef_), feature_names, model.coef_)
    )

# Create a DataFrame
coefficients_df = pd.DataFrame(coefficients_data, columns=["Player", "Variable", "Coefficient"])

# Plot the coefficients comparison
plt.figure(figsize=(14, 6))
sns.barplot(data=coefficients_df, x="Variable", y="Coefficient", hue="Player")
plt.title("Comparison of Coefficients for Luka Doncic and LeBron James")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
